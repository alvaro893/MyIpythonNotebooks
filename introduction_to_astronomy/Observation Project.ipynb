```python
# DATA
def toDecimal(ang, minutes):
    return (ang*60 + minutes)/60
# sun data
sunangle = [toDecimal(45,40),toDecimal(45,48),toDecimal(45,57),toDecimal(46,2),toDecimal(46,7),toDecimal(46,11),toDecimal(46,14),
            toDecimal(46,16),toDecimal(46,16),toDecimal(46,16),toDecimal(46,14),toDecimal(46,12),toDecimal(46,8),toDecimal(46,3),
            toDecimal(45,56),toDecimal(45,50)]
sunazimuth = [toDecimal(168,39),toDecimal(170,22),toDecimal(172,6),toDecimal(173,50),toDecimal(175,34),toDecimal(177,18),
              toDecimal(179,3),toDecimal(180,47),toDecimal(182,32),toDecimal(184,16),toDecimal(186,0),toDecimal(187,44),
              toDecimal(189,28),toDecimal(191,11),toDecimal(192,54),toDecimal(194,37),]

# time to minutes
time = array(['2015-05-07 12:45:00','2015-05-07 12:50:00','2015-05-07 12:55:00','2015-05-07 13:00:00','2015-05-07 13:05:00',
              '2015-05-07 13:10:00','2015-05-07 13:15:00','2015-05-07 13:20:00','2015-05-07 13:25:00','2015-05-07 13:30:00',
              '2015-05-07 13:35:00','2015-05-07 13:40:00','2015-05-07 13:45:00','2015-05-07 13:50:00','2015-05-07 13:55:00',
              '2015-05-07 14:00:00'], dtype='datetime64[m]')
```

```python
# Plot sun altitude
plot(time, sunangle, '.')

title("Sun's elevation angle vs Time")
xlabel('Time in minutes')
ylabel('Angle in decimal degrees')
```

```python
# Plot sun azimuth
plot(time, sunazimuth, '.')
title("Azimuth vs Time")
xlabel('Time in minutes')
ylabel('Angle in decimal degrees')
```
