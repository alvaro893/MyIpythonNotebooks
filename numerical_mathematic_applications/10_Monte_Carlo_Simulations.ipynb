# Introduction

In many situations we are not able to construct an analytic model to explain the behavior of the observed system due to its complexity or difficulty of the model. In some circustances, it may not even be feasible to observe the behavior directly or to conduct experiments. Even further, the system we want to test may not even exist yet! In these instances the simulation may give indirect information about the system and help to improve the design. (Giordiano, Weir & fox, 2003, p. 175-176)

Monte Carlo simulations are algorithms that rely on repeated random sampling to obtain numerical results (Wikipedia, 2014).  Typically simulations are run several times in order to obtain statistical results of the object. The advantage of the Monte Carlo simulations is the relative ease to approximate complex or difficult probabilistic systems.

This week we practise simple Monte Carlo simulations. In the first problem we approximate the area under a curve.

```python
## To do 1 & 2
# my algorithm of Aproximation of pi
def makePi(n):
    # number of values inside circle
    N = 0
    # generate random values
    x = rand(n)
    y = rand(n)
    
    # share of values inside the circle
    x_in = []
    y_in = []
    
    # function which calculates a quarter of circumference
    xc = linspace(0, 1, 100)
    yc = lambda(x): sqrt(1 - x**2)
    
    # count numbers inside the circle
    for i in range(n):
        if (yc(x[i]) > y[i]):
            N = N + 1
            x_in.append(x[i])
            y_in.append(y[i])
            
    # from area of circle formula 'A = pi/4*radius' where n is the total area and N/n the area inside the circle.
    # Note the division by four because we use a quarter of a circle
    PI = N/n * 4
    print "pi =", PI
    print "difference:", abs(pi - PI) * 100,"%" 
    
    plot(xc, yc(xc))
    plot(x, y, '.')
    plot(x_in, y_in, '.')
    title("n = {0}".format(n), fontsize = 16)

makePi(10)
```

```python
#
def makeSph(n):
    # number of values inside sphere
    N = 0
    
    # generate random values
    x = rand(n)
    y = rand(n)
    z = rand(n)
    
    # share of values inside the circle
    x_in = []
    y_in = []
    z_in = []
    
    # function which calculates a Sphere
    xs = linspace(0, 1, 100)
    f_sphz = lambda x,y: sqrt(1 - x**2 -y**2)
        
    # count numbers inside the sphere and to record their position
    for i in range(n):
        if (f_sphz(x[i], y[i]) > z[i] ):
            N = N + 1
            x_in.append(x[i])
            y_in.append(y[i])
            z_in.append(z[i])
            
    # 
    # Note that everything occurs in the first octant so in order to get the whole volume we need to multiply by 8
    Vs = N/n*8  # N/n is 1/8 from total volume
    print "volume =", Vs
    print "difference:", abs(Vs - 4/3*pi) * 100,"%" 
    
    # ploting random points in first octant
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.plot(x, y, z, 'o')
    ax.plot(x_in, y_in, z_in, 'o')
    #ax.plot(xs, xs, f_sphz(xs,xs))
    xlabel('x')
    ylabel('y')
    
    
makeSph(100)
```
