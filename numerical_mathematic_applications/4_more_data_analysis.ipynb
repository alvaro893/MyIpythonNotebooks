## Problem 1 - Simple pendulum

Adapted from: [physics.info](http://physics.info/)   

In this experiment, simple pendulums of different lengths were constructed and their periods measured. This data set looks linear, but is it really? What effect does length have on the period of a simple pendulum?

Data: [simple-pendulum.txt](http://physics.info/curve-fitting/simple-pendulum.txt)

```python
# Read the data into Python
link_name = 'http://physics.info/curve-fitting/simple-pendulum.txt'
pendulum =  pd.read_csv(link_name, '\t', header = [0,1])

# Display 4 first rows
pendulum[:5]
```

```python
# Get the columns from the data set
Length = pendulum['Length'].astype(int32)
Period = pendulum['Period'].astype(float)

# Plot the Period versus Length
figure()
plot(Length, Period, 'o')
title('Period (s) versus Length (cm)')
```

```python
# this converts data in array type to use in polyfit
le = np.squeeze(np.asarray(Length.values))
pe = np.squeeze(np.asarray(Period.values))

# Trying to fit data with Polyfit with degree = 1
p = polyfit(le, pe, 1)
period_fit1 = polyval(p, le)

#plot results
plot(le, period_fit1)
plot(Length, Period, 'o')
title("fit with linear function")
xlabel("length L (cm)",fontsize=18)
ylabel("Period T (s)",fontsize=18)
```

Using the Period of oscilation for pendulum formula we can fix it in this way, and get a better fit $$T \approx 2\pi \sqrt\frac{L}{g} \quad \rightarrow \quad T^{2} \approx (2\pi)^{2} \frac{L}{g} \quad \rightarrow \quad T^{2} \approx kL$$


```python
# Using polyfit with T**2
p = polyfit(le, pe ** 2, 1)
period_fit2 = polyval(p, le)

#plot results
plot(le, period_fit2)
plot(Length, Period ** 2, 'o')
title("fit with fix function")
xlabel("length L (cm)",fontsize=18)
ylabel("Period $T^{2}$ ($s^{2}$)",fontsize=18)
```

The plot looks better than the previous one, we can use the error average formula to determine which one is closer to solution, $ e_{mse} = \frac{1}{N}\sum_{n=0}^{N-1} abs\left( y[n] - y_{fit}[n]\right) $

```python
# Average error of lineal fit
e1 = 0
for n in range(len(period_fit1) - 1):
    e1 = e1 + abs((pe[n] - period_fit1[n]))
e1 = (e1)/len(pe)
print "e1:",e1

# Average error of fix pendulum formula fit
e2 = 0
for n in range(len(period_fit2) - 1):
    e2 = e2 + abs((pe[n] - period_fit2[n]))
e2 = (e2)/len(pe)
print "e2:",e2
```

As we can see the second fit is pretty better because is closer to one

  
## Problem 2 - Vostok

Adapted from: [physics.info](http://physics.info/)    

Snow rarely gets a chance to melt in [Antarctica](http://en.wikipedia.org/wiki/Antarctica), even in the summer when the sun never sets. In the interior of the continent, the temperature of the air hasn't been above the freezing point of water in any significant way for the last 900,000 years. The snow that falls there accumulates and accumulates and accumulates until it compresses into rock solid ice — up to 4.5 km thick in some regions.    
![rock solid ice](http://upload.wikimedia.org/wikipedia/commons/thumb/8/8f/Fryxellsee_Opt.jpg/330px-Fryxellsee_Opt.jpg)   
Since the snow that falls is originally fluffy with air, the ice that eventually forms still holds remnants of this air — very, very old air. By examining the isotopic composition of the gases in carefully extracted cores of this ice we can learn things about the past climatic conditions on earth. By extension we might also predict some things about the climate of the future. The columns in this data set are as follows:

- Age of air (years before present)
- Temperature anomaly with respect to the mean recent time value (℃)
- Carbon dioxide concentration (ppm)
- Dust concentration (ppm)

Source: Adapted from [Petit, et al. 1999](http://www.ncdc.noaa.gov/paleo/icecore/antarctica/vostok/vostok_data.html).

```python
# Read the data into Python
link_name = 'http://physics.info/curve-fitting/vostok.txt'
vostok =  pd.read_csv(link_name, '\t', header = [0,1])

# Display 4 first rows
vostok[:5]
```

```python
# Get the columns from the data set
age = vostok['age'].astype(int32)
dt = vostok['∆T'].astype(float)

# Plot the dT versus age
figure()
plot(age, dt)
title('$\Delta T$ versus age')
```

```python
# Get the co2 column from the data set
co2 = vostok['CO2'].astype(float)

# A set of overlapping time series graphs for CO2 concentration and temperature anomaly
fig, ax1 = plt.subplots(figsize=(17,6))

ax1.plot(age, co2, lw=2, color="blue")
ax1.set_ylabel(r"CO$_{2}$", fontsize=18, color="blue")
ax1.set_xlabel("Age", fontsize=18)
ax1.set_title("CO$_{2}$ and $\Delta T$ vs Age",fontsize=18)
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(age, dt, lw=2, color="red") 
ax2.set_ylabel(r"$\Delta T$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

```python
# Scatter plot of temperature anomaly vs. CO2 concentration.

figure()
plot(co2, dt, 'oc')
title('$\Delta T$ versus $CO_{2}$', fontsize = 18)
xlabel("CO$_{2}$", fontsize = 18)
ylabel("$\Delta T$", fontsize = 18)
```

#### How are atmospheric carbon dioxide concentration and temperature anomaly related?
Generally it seems that more temperature implies more carbon dioxide concentration so they are directly related.

#### What temperature anomaly might one expect given current atmospheric CO2 levels?
If we do a zoom out on plot (look below) in our present (0 ~ 3000 age) , we can see that in the relative present we have close to 285 ppm and a Temperature anomaly between -2 and 0.5 degrees centigrates.

```python
fig, ax1 = plt.subplots(figsize=(6,5))

ax1.plot(age[:100], co2[:100], lw=2, color="blue")
ax1.set_ylabel(r"CO$_{2}$", fontsize=18, color="blue")
ax1.set_xlabel("Age", fontsize=18)
ax1.set_title("CO$_{2}$ and $\Delta T$ vs Age (zoom in)",fontsize=18)
ax1.set_xlim(0,4000)
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(age[:100], dt[:100], color="red") 
ax2.set_ylabel(r"$\Delta T$", fontsize=18, color="red")
ax2.set_xlim(0,4000)
for label in ax2.get_yticklabels():
    label.set_color("red")
```

```python
# Get the dust column from the data set
dust = vostok['dust'].astype(float)

# a set of overlapping time series graphs for dust and temperature anomaly.
fig, ax1 = plt.subplots(figsize=(17,6))

ax1.plot(age, dust, lw=2, color="blue")
ax1.set_ylabel(r"Dust", fontsize=18, color="blue")
ax1.set_xlabel("Age", fontsize=18)
ax1.set_title("Dust and $\Delta T$ vs Age",fontsize=18)
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(age, dt, lw=2, color="red") 
ax2.set_ylabel(r"$\Delta T$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

```python
# Scatter plot of temperature anomaly vs. dust concentration

figure()
plot(dust, dt, 'oc')
title('$\Delta T$ versus Dust', fontsize = 18)
xlabel("Dust", fontsize = 18)
ylabel("$\Delta T$", fontsize = 18)
```

####How are atmospheric dust concentration and temperature anomaly related?
As we can see on the overlapping graph they seem widely inverse related, in other words when there are maximum dust concentrate there are minimum temperature anomaly.

####What global average temperature anomaly might one expect from exceptionally high levels of atmospheric dust?
On the scatter plot it is seen clearly: The lower values of temperature anomaly, around -8 degrees celsius
