## To do

- Use Python to make a graph of the following function.   
- Find graphically approximate solutions for the equations.   
- Create a script that finds an approximate solution iteratively and find solution for the equations.

  (a) $ f(x) = 2x^4 - x^3 + x^2 - 10 = 0 $   
  (b) $ f(t) = 2e^{-3t}-t^2 = 0 $   
  (c) $  f(t) = t-sin(t) = 0 $


```python
from __future__ import division
%pylab inline
```

```python

x = linspace(-10, 10, 100000)
ya = 2*x**4 - x**3 + x**2 - 10

# stright line in 0
y0 = x*0

plot(x,ya)
plot(x,y0)
xlabel('x',fontsize=18)
ylabel('y',fontsize=18)
title("$ f(x) = 2x^4 - x^3 + x^2 - 10 = 0 $", fontsize=18)
grid()
xlim((-2, 2))
ylim((-20, 10))
```

graphically we can see the solutions looking at the intersections, there are a value between -1.5 and -1.0 and another value close to 1.5

```python
t = linspace(-10, 10, 1000)
yb = 2*exp(-3*t) - t**2
y0 = t*0

plot(t,yb)
plot(t,y0)
xlabel('t',fontsize=18)
ylabel('y',fontsize=18)
grid()
title("$ f(t) = 2e^{-3t}-t^2 $", fontsize=18)
ylim(-10, 10)
xlim(-2, 2)
```

In this case the equation reaches zero when 't' is something more of 0.5

```python
yc = t - sin(t)

plot(t,yc)
plot(t,y0)
xlabel('t',fontsize=18)
ylabel('y',fontsize=18)
grid()
title("$  f(t) = t-sin(t) $", fontsize=18)
ylim(-10, 10)
xlim(-5, 5)
```

In this equation it is easy to see that '0' is the only solution

```python
# script that finds an approximate solution iteratively
# assuming '0' as right side
# this was my solution
def solv(x,y):
    a = [] # array to keep index
    for n in range(len(y)):
        if(y[n] < 0.1 and y[n] > -0.1):
            a.append(n)
    sol_index = a[(int)(len(a)/2)]
    return x[sol_index]

print "first solution (ineficient)"  
print "solution for a",solv(x,ya)
print "solution for b",solv(t,yb)
print "solution for c",solv(t,yc)
print '\n'

# but this function is not able to find more than one solution
# and it is too much ineficient

# this is the Sakari's solution which is really simple and effective
def solv(x,y):
    index = y > 0
    return x[diff(index)]

print "second solution (nice)"
print "solution for a",solv(x,ya)
print "solution for b",solv(t,yb)
print "solution for c",solv(t,yc)
```

## To do

- Use the Newton-Raphson technique to find the value of a root of the following equations correct to 7 decimals.   
- Try different initial values. Do they all converge to the same solution?    
- How many iterations each of the solution takes?   
- Make a plot of the iteration result versus iteration cycle.   

 (a) $ 2cos(x) = x^2, \quad x_1 = 0.8 $   
 (b) $ 3x^3 - 4x^2 + 2x - 9 = 0, \quad x_1 = 2 $   
 (c) $ e^{x/2} - 5x = 0, \quad x_1 = 6 $   
 (d) $ ln(x) = \dfrac{1}{x}, \quad x_1 = 1.6 $   
 (e) $ \sin(x) + \dfrac{2x}{\pi} = 1, \quad x_1 = 0.6 $


```python
# First I create a function which calculates the numeric derivate
# using the derivate definition: http://en.wikipedia.org/wiki/Derivative#Rigorous_definition
# by taking the function to derivate as argument "f" and the point "x1"
# where the function is derivated
def deriv(f,x1):
    h = 1e-8 # h must be close to zero but not zero.
    return (f(x1 + h) - f(x1) ) / h

# This algorithm performs the Newton-Raphson method by looping
# until a seven decimal point aproximation is reached,
# "f" is the equation to solve previously cleared to 0
# "x" is the arbitrary value where the algorithm starts
def newton(f, x):
    iterations = 0
    iteration_results = [x]
    while(not( -1e-7 < f(x) < 1e-7)):
        x = x - f(x)/deriv(f,x)
        iteration_results.append(x)
        iterations += 1
        
        #number of iterations limited to 1000 just in case
        if(iterations >= 1000): 
            return "solution not reached"
    
    #plot iterations vs iteration cycle    
    plot(range(iterations+1), iteration_results, '-o')
    xlabel("number of iterations")
    ylabel("result of iteration")
    grid()
    
    print "number of iterations:", iterations
    return x

# Here are the equations with the right side cleared to zero
# and represented by functions
def y_a(x):
    return 2*cos(x) - x**2
def y_b(x):
    return 3*x**3 - 4*x**2 + 2*x - 9
def y_c(x):
    return exp(x/2) - 5*x
def y_d(x):
    return log(x) - 1/x
def y_e(x):
    return sin(x) + 2*x/pi - 1
```

```python
# Here are the soluctions and cheking out for 7 decimals
# the solution should return a zero when we use it the 
# original function

#a
sol_a = newton(y_a, 0.8)
print "solution a:", sol_a
print "checking a:", -1e-7 < y_a(sol_a) < 1e-7
```

```python
#b
sol_b = newton(y_b, 2)
print "solution b:", sol_b
print "checking b:", -1e-7 < y_b(sol_b) < 1e-7
```

```python
#c
sol_c = newton(y_c, 6)
print "solution c:", sol_c
print "checking c:", -1e-7 < y_c(sol_c) < 1e-7
```

```python
#d
sol_d = newton(y_d, 1.6)
print "solution d:", sol_d
print "checking d:", -1e-7 < y_d(sol_d) < 1e-7
```

```python
#e
sol_e = newton(y_e, 0.6)
print "solution e:", sol_e
print "checking e:", -1e-7 < y_e(sol_e) < 1e-7
```
