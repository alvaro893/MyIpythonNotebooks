## To do 2

- Create histograms of the values for the '∆T', 'CO2' and 'dust' variables.
- How many percent of the values are below the mean and above the mean?
- How many percent of the values are between $ \mu - sigma $ and $\mu + \sigma $, where $\mu$ is the mean value and $\sigma$ is the standard deviation?  

See also: [histogram demo](http://matplotlib.org/1.2.1/examples/api/histogram_demo.html), [Pandas missing data](http://pandas.pydata.org/pandas-docs/version/0.13.1/10min.html#missing-data), [Panda vieving data](http://pandas.pydata.org/pandas-docs/version/0.13.1/10min.html#viewing-data), [flatten()](http://docs.scipy.org/doc/numpy/reference/generated/numpy.ndarray.flatten.html)

```python
import pylab as P

# in order to write less code I created this function
def calculateValues(data, title):
    P.figure(figsize=(14,5))
    x = data.dropna().values.flatten()
    res = P.hist(x, 50, normed = True)
    P.xlabel("$\mu - \sigma$, $\mu$, $\mu +\sigma$ respectively with each vertical line ", fontsize = 18)
    
    # This draws the mean, and the standard deviations on the histogram
    P.axvline(x.mean(), color='r', linestyle='dashed', linewidth=2)
    P.axvline(x.mean() - x.std(), color='g', linestyle='dashed', linewidth=2)
    P.axvline(x.mean() + x.std(), color='g', linestyle='dashed', linewidth=2)
    
    P.title(title, fontsize = 18)
    
    # This count the ocurrences above and below of the mean
    xlow = (x < x.mean()).tolist().count(True)
    xhigh = (x > x.mean()).tolist().count(True)
    
    # test purposes
    print "checking we have all values...", xlow + xhigh , xlow + xhigh == len(x), "\n"
    
    # Now, we can calculate the percent
    print "values below the mean: ", xlow/len(x) * 100, "%"
    print "values above the mean: ", xhigh/len(x) * 100, "%", "\n"
    
    # Percent of values above and below of (mean - standard deviation)
    xlow = (x < (x.mean() - std(x))).tolist().count(True)
    xhigh = (x > (x.mean() - std(x))).tolist().count(True)
    print "values below the mean - std : ", xlow/len(x) * 100, "%"
    print "values above the mean - std: ", xhigh/len(x) * 100, "%", "\n"
    
    # Percent of values above and below of (mean + standard deviation)
    xlow = (x < (x.mean() + std(x))).tolist().count(True)
    xhigh = (x > (x.mean() + std(x))).tolist().count(True)
    print "values below the mean + std : ", xlow/len(x) * 100, "%"
    print "values above the mean + std: ", xhigh/len(x) * 100, "%", "\n"
```

```python
calculateValues(temp, "$\Delta T$ histogram with mean line")
```

```python
calculateValues(co2, "$CO_{2}$ histogram with mean line")
```

```python
calculateValues(dust,"Dust histogram with mean line")
```
