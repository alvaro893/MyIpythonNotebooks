# Introduction

Many functions cannot be integrated analytically, but can be evaluated [numerically](http://en.wikipedia.org/wiki/Numerical_integration). More important, numerical integration is used to solve differential equations. For example, the relationship between the speed and travelled distance is expressed in differential form $$ v(t) = \dfrac{dx(t)}{dt} $$
where 

- $ v(t) $ is the velocity,   
- $ s(t) $ is the travelled distance, and
- $ t $ is time.

If we solve for the distance, the equation transforms to integral form
$$ s(t) = \int_{t_0}^{t_1}v(t) dt $$

Sometimes, we even don't know the function analytically, but the integrand may be known only at certain points. Then we need to work with the samples of the function
$$ f[n] = \left \{ f_0, f_1, f_2, ..., f_n \right \} $$ 

one of the simplest algorithm to estimate the numerical integral is [Trapezium rule](http://en.wikipedia.org/wiki/Trapezoidal_rule)
$$ \int_{a}^{b} f(x)\, dx \approx \frac{h}{2} \sum_{k=1}^{N} \left( f(x_{k+1}) + f(x_{k}) \right){}= \frac{h}{2}(f(x_1) + 2f(x_2) + 2f(x_3) + \ldots + 2f(x_N) + f(x_{N+1}))$$
where

- $ h = \dfrac{b-a}{N} $ is the strip width
- $ N $ is the number of strips

```python
%pylab inline
from __future__ import division

S = r_[5, 10, 20, 30, 50, 60, 70, 90, 100]
e = r_[0, 19, 57, 94, 173, 216, 256, 343, 390]
```

Using the interpolation formula we can approximate unknown 'e(S)' values with closer known 'e' values $$e = e_1 + \frac{S-S_1}{S_2-S_1}(e_2-e_1)$$

```python
# To do 1
# the closer values to S = 40 are between 30 and 50.
e40_inter = e[3] + (40 - S[3])/(S[4] - S[3]) * (e[4] - e[3])
print "e(S=40) = ", e40_inter

# the closer values to S = 80 are between 70 and 90.
e80_inter = e[6] + (80 - S[6])/(S[7] - S[6]) * (e[7] - e[6])
print "e(S=80) = ", e80_inter
```

```python
# To do 2

#Using polyfit we can find a model for the data and predict values
p = polyfit(S, e, 1)
print "p:", p

e40_fit = 40*p[0]+p[1]
e80_fit = 80*p[0]+p[1]
print "model Estimation"
print "e (S=40) =", e40_fit
print "e (S=80) =", e80_fit

#
e_fit = polyval(p, S)

plot(S, e, 'o', label = "data")
plot(S, e_fit,  label = "model")
title("S versus e", fontsize = 18)
xlabel("Stress (in/in)", fontsize = 18)
ylabel("elongation (lb/in$^{2}$)", fontsize = 18);
legend(loc=2);
```

```python
# Numerical solution

# e = c1*S model: derivate(relative to c1) of e_rms result
c1 = sum(e*S)/sum(S**2)


# Trying the model
print c1
print "c1*S1 = e:", c1*S, '\n'
print "e:",e

# how much is the error?
print "error in each data value:", c1*S - e

# Plotting both results of c1 and comparing with polyfit solution
plot(e, S, 'o', label = "data")
plot(c1*S, S, label = "model")
plot(e_fit, S, label = "data fit (polyfit)")
title("S versus e", fontsize = 18)
xlabel("Stress (in/in)", fontsize = 18)
ylabel("elongation (lb/in$^{2}$)", fontsize = 18);
legend(loc=2);
```

```python
# values of e when S=40 and S=80 whith this model
e40_numeric = c1*40
e80_numeric = c1*80
print "values of e when S=40 and S=80 whith this model (e = c1*S)"
print e40_numeric, " ", e80_numeric
print "the same with interpolation"
print e40_inter, " ", e80_inter
print "And with polyfit"
print e40_fit, " ", e80_fit
```

Therefore we can observe that the interpolation is a better result than the model $e = c_{1}S$ because is closer to the data fit (with polyfit)

```python
# Function which performs a linear interpolation in two value sets
def interlin(x_value,x,y):
    a,b,i = 0,0,0 # a and b perform the closest values to the desired value
    if(x_value == x.any()): # just in case the value exists already
        print "value exists in set"
        return
    for i in range(0, x.size):
        if(x[i] > x_value):
            b = i
            a = i-1
            break
    # this is the linear interpolation formula
    y_value = y[a] + (y[b] - y[a]) * (x_value - x[a])/(x[b] - x[a]) 
    #print y[a],y[b]
    #print "test:", y[a] < y_value < y[b] 
    return y_value

# y-values for x = 22, 31, 32, 37, and 38
values = (22, 31, 32, 37, 38)

for i in values:
    print "x = ", i, " y =", interlin(i, x, y)
```

```python
# To do 2
p = polyfit(x,y,2)
print p

# pol is the polynomial but performing as a function
pol = lambda(x): x**2*p[0] + x*p[1] +p[2]

for i in values:
    print "x = ", i, " y =", pol(i)
```

```python
# comparison
print "difference"
print "x = 22 ->", abs(interlin(22, x, y) - pol(22))
print "x = 31 ->", abs(interlin(31, x, y) - pol(31))
print "x = 32 ->", abs(interlin(32, x, y) - pol(32))
print "x = 37 ->", abs(interlin(37, x, y) - pol(37))
print "x = 38 ->", abs(interlin(38, x, y) - pol(38))
```

The second order polynomial approximation is better because the linear interpolation is less accurate due to linear interpolation is assuming that the curve is linear

The total distance travelled by the biker: 
In order to get the travelled space (s) we have to integrate the function which describes the speed (v) $$s = \int_{t_0}^{t_1}v(t)\mathrm{d} t$$

```python
# My implementation of the trapecium algorithm
# f:function to integrate, [a,b]: interval, N: number of strips
def trapecium(f, a, b, N):
    h = (b - a)/N
    s = h/2*(f(a) + f(b))
    for n in range(1,N):
        x = a + n*h
        s = s + f(x)*h
    return s

s = trapecium(new_Speed, Time[0], Time[-1], 1000)
print "distance travelled:", s, "meters or", s/1000, "kilometers"
```
