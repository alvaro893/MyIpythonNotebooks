Plotting a $$y=k$$
in this case $k$ is $b$ the origin in $y$ and the slope is $0$

```python
b = 4
k = 0
x1 = arange(0,10.1,0.1)
y1 = k*x1+b
grid('on')
title("$y=k$",fontsize=18)
plot(x1,y1)

```

```python
b = 0
k = 1
x2 = arange(0,10.1,0.1)
y2 = k*x2+b
title("$y=kx$",fontsize=18)
plot(x2,y2)
grid('on')
```

```python

k = 1
x3 = arange(0,10,0.1)
y3 = k/x3+b
grid('on')
title("$y=k/x$",fontsize=18)
plot(x3,y3)
```

```python
k = 1
x = arange(0,20,0.1)
y = k*x**2
grid('on')
title("$y=kx^2$",fontsize=18)
plot(x,y)
```

```python
k = 1
x = arange(0,20,0.1)
y = k*sqrt(x)
grid('on')
title("$y=k\sqrt{x}$",fontsize=18)
plot(x,y)
```

```python
n = 3
k = 1
x = arange(-10,10,0.1)
y = k*x**n
grid('on')
title("$y=kx^n$", fontsize=18)
plot(x,y)
```

```python
b = 5
k = 1
x = arange(-10,10,0.1)
y = k*x+b
plot(x,y)
title("$y=kx+b$",fontsize=18)
grid('on')
```

```python
b = 5
a = 2
c = -1
x = arange(-10,10,0.1)
y = a*x**2+b*x+c
plot(x,y)
title("$y=k$",fontsize=18)
grid('on')
```

```python
a = 2
n = 3
k = 2
x = arange(0,10,0.1)
y = a*n**(x*k)
title("$y=an^{kx}$",fontsize=18)
plot(x,y)
grid('on')
```

```python
a = 2
n = 3
k = 2
x = arange(0,10,0.1)
y = a*n**(-x*k)
title("$y=an^{-kx}$",fontsize=18)
plot(x,y)
grid('on')
```

```python
a = 2
n = 3
k = 2
x = arange(-10,10,0.1)
y = a*(1-n**(-x*k)+b)
title("$y=a(1-n^{-kx})+b$",fontsize=18)
plot(x,y)
grid('on')
```

```python
a = 2
b = 3
k = 2
x = arange(-2,2,0.1)
y = a*sin(k*x+b)
title("$y=asin(kx+b)$",fontsize=18)
plot(x,y)
grid('on')
```
