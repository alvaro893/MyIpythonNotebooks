#Problem 3. Manual fit

- Use the trial-and-error method to manually find the values for the parameters giving the best fit of the model for each of the case.
- Show graphically the steps how your fit is getting better (or worse!).

```python
# Example how to match graphically the linear model. First match.

x = linspace(0.0, 3.0, 50)
y1 = 2.0*x - 1.0

plot(data['x'], data['z'], 'o')
plot(x, y1, '-')
title('$ y = 2.0x - 1.0 $')
```

```python
# Let's do a better match.

y2 = 2.5*x + 0.0
plot(data['x'], data['z'], 'o')
plot(x, y2, '-')
title('$ y = 2.5x + 0.0 $')
```

```python
y3 = 3*x + 0.0
plot(data['x'], data['z'], 'o')
plot(x, y3, '-')
title('$ y = 3x + 0.0 $')
```

```python
y4 = 3.25*x + 0.0
plot(data['x'], data['z'], 'o')
plot(x, y4, '-')
title('$ y = 3.25x + 0.0 $')
```

```python
# Combine both matches on the same graph

plot(data['x'], data['z'], 'o')
plot(x, y1, '-')
plot(x, y2, '-')
plot(x, y3, '-')
plot(x, y4, '-')
```

```python
# square root model

y1 = 2.0*sqrt(x) 

plot(data['x'], data['u'], 'o')
plot(x, y1, '-')
title('$ y = 2.0\sqrt{x}$')
```

```python
y2 = 1.0*sqrt(x)

plot(data['x'], data['u'], 'o')
plot(x, y2, '-')
title('$ y2 = 1.0\sqrt{x} $')
```

```python
plot(data['x'], data['u'], 'o')
plot(x, y1, '-')
plot(x, y2, '-')
```

```python
# exponential model

y1 = 1.0*exp(x) 

plot(data['x'], data['v'], 'o')
plot(x, y1, '-')
title('$ y1 = 1.0e^{x} $')
```

```python
# quadratic model

y1 = -2*x**(2)+3*x+0.0

plot(data['x'], data['y'], 'o')
plot(x, y1, '-')
```

```python
y2 = -1*x**(2)+3*x+0.0

plot(data['x'], data['y'], 'o')
plot(x, y2, '-')
```

```python
y3 = -1*x**(2)+3*x-1

plot(data['x'], data['y'], 'o')
plot(x, y3, '-')
```

```python
k=0.5
y4 = -k*x**(2)+3*k*x-1*k

plot(data['x'], data['y'], 'o')
plot(x, y4, '-')
```

```python
k = 0.5
c = 0.05
y5 = -k*x**(2)+3*k*x-c

plot(data['x'], data['y'], 'o')
plot(x, y5, '-')
```

```python
plot(data['x'], data['y'], 'o')
plot(x, y1, '-')
plot(x, y2, '-')
plot(x, y3, '-')
plot(x, y4, '-')
plot(x, y5, '-')
```

#Problem 4. Least squares polynomial fit

###To do
- Play with some of the examples given in the polyfit() documentation to get familiar how the functions work.
- Use the polyfit() function to find the best polynomial fit for each of the cases given in the previous problem.
- How much do the values given by the model differ from the given data values?

```python
# Example how to fit a polynomial function to the given data extracted with pandas' csv_read.

x_val = data['x']
y_val = data['z']
p = polyfit(x_val, y_val, 1)
print p
y_est = polyval(p, x_val)
plot(x_val, y_val, 'o')
plot(x_val, y_est, '-')
```

```python
x_val = data['x']
y_val = data['u']
p = polyfit(x_val, y_val, 2)
print p
y_est = polyval(p, x_val)
plot(x_val, y_val, 'o')
plot(x_val, y_est, '-')
```

```python
x_val = data['x']
y_val = data['v']
p = polyfit(x_val, y_val, 2)
print p
y_est = polyval(p, x_val)
plot(x_val, y_val, 'o')
plot(x_val, y_est, '-')
```

```python
x_val = data['x']
y_val = data['y']
p = polyfit(x_val, y_val, 2)
print p
y_est = polyval(p, x_val)
plot(x_val, y_val, 'o')
plot(x_val, y_est, '-')
```

```python
x_val = data['x']
y_val = data['e']
p = polyfit(x_val, y_val, 20)
print p
y_est = polyval(p, x_val)
plot(x_val, y_val, 'o')
plot(x_val, y_est, '-')
```
