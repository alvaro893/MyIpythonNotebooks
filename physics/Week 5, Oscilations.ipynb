## 2 Resonant Frequency
The resonant frequency is the frequency which the energy keeps constant on the RLC circuit

```python
Vp = 1
L = 8.2e-3
C = 330e-6
R = 10
w0 = 1 / sqrt(L*C)
print "resonance frequency:",ceil(w0), "rad/s"
```

## 3 What is the damping coefficient for the exponential decay?
As any real circuit has resistance in the wire there are always a certain damping. The damping coeficient for a RLC circuit is $2L/R$ As the resistance increases, oscillations decay more rapidly


```python
t = 2*L/R
t
```

## 4 Q Factor
The Q factor is a parameter that indicates the rate of loss of energy. A Big Q factor means that the system loss energy slowly. In a RLC circuit can measured by $Q=\omega_0 L/R$
### Resonance curves

```python
def Ip(w,R): return Vp/sqrt(R**2 + (w*L - 1/(w*C))**2)
w = linspace(0,1700,1000)

matplotlib.rcParams.update({'font.size': 13})
title("Q factor", fontsize = 17)
xlabel("$\omega$: Angular frequency (rad/s)")
ylabel("Ip: Current peak (A)")

Q = w0*L/(R/2)
plot(w,Ip(w,R/2), label = "Q = " + str(around(Q,2))) # 1/2 R

Q = w0*L/(R/1)
plot(w,Ip(w,R), label = "Q = " + str(around(Q,2))) # R

Q = w0*L/(2*R)
plot(w,Ip(w,2*R), label = "Q = " + str(around(Q,2))) # 2R

legend(loc=0)
```
